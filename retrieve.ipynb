```python
import pandas as pd
```

```python
df = pd.read_csv("results/literature_review_results.csv")
df["year"] = pd.to_numeric(df["year"], errors="coerce")
df["citation_count"] = pd.to_numeric(df["citation_count"], errors="coerce")
df_filtered = df[(df["year"] >= 2023) & (df["citation_count"] > 10)]
print(df_filtered[["title", "citation_count", "url"]].head(20).to_string(index=False))
```
